```python
import torch
import torch.nn as nn
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

wf, whf, bf = 0.5, 0.1, 0
wi, whi, bi = 0.6, 0.2, 0
wc, whc, bc = 0.7, 0.3, 0
wo, who, bo = 0.8, 0.4, 0

h, c = 0, 0
x_sequence = [1, 2, 3]

print("Calculation Steps (Matching PDF):")
for t, x in enumerate(x_sequence):
    f = sigmoid(wf * x + whf * h + bf)
    i = sigmoid(wi * x + whi * h + bi)
    c_tilde = np.tanh(wc * x + whc * h + bc)
    c = (f * c) + (i * c_tilde)
    o = sigmoid(wo * x + who * h + bo)
    h = o * np.tanh(c)
    print(f"Step {t+1}: h = {h:.3f}, c = {c:.3f}")

f4 = sigmoid(wf * 4 + whf * h + bf)
print(f"\nForget Gate for x4 (f4): {f4:.3f}")
```

```python
model = nn.LSTM(input_size=1, hidden_size=1, batch_first=True)

with torch.no_grad():
    model.weight_ih_l0[:] = torch.tensor([[0.6], [0.5], [0.7], [0.8]])
    model.weight_hh_l0[:] = torch.tensor([[0.2], [0.1], [0.3], [0.4]])
    model.bias_ih_l0[:] = torch.tensor([0.0, 0.0, 0.0, 0.0])
    model.bias_hh_l0[:] = torch.tensor([0.0, 0.0, 0.0, 0.0])

inputs = torch.tensor([1.0, 2.0, 3.0]).view(1, 3, 1)
out, (hn, cn) = model(inputs)

print(f"PyTorch Final Hidden State: {hn.item():.3f}")
```
